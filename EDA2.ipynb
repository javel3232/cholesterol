#**Choresterol**

El conjunto de datos de enfermedades cardíacas es una colección que aborda el diagnóstico de enfermedades cardíacas mediante el análisis de varios atributos médicos. Recopilado de cuatro instituciones médicas distintas, este conjunto de datos contiene información de pacientes de diversas ubicaciones, incluidos Cleveland, Budapest, Long Beach (California) y Zurich (Suiza). Con un total de 76 atributos, se ha utilizado una selección de 14 de ellos en investigaciones anteriores. El objetivo principal de estas investigaciones ha sido predecir la presencia de enfermedades cardíacas en los pacientes, clasificando si están presentes o ausentes. La tarea se ha centrado en distinguir entre la presencia y ausencia de enfermedades cardíacas, con la presencia codificada en valores del 1 al 4 y la ausencia codificada como 0. Es importante destacar que, aunque se han realizado estudios previos con este conjunto de datos, la mayoría de ellos se han centrado en la base de datos de Cleveland. Este conjunto de datos ha sido ampliamente utilizado por investigadores de aprendizaje automático y se ha investigado su aplicabilidad para la predicción de enfermedades cardíacas.

### **Descripción de las variables del conjunto de datos:**

| Atributo   | Descripción                                                                                   |
|------------|-----------------------------------------------------------------------------------------------|
| **Age**        | Edad del paciente en años.                                                                   |
| **Sex**        | Sexo del paciente. 1 = masculino, 0 = femenino.                                             |
| **CP**         | Tipo de dolor en el pecho.                                                                   |
|            | - Valor 1: Angina típica.                                                                    |
|            | - Valor 2: Angina atípica.                                                                   |
|            | - Valor 3: Dolor no anginal.                                                                 |
|            | - Valor 4: Asintomático.                                                                     |
| **Trestbps**   | Presión arterial en reposo en mm Hg al ingreso al hospital.                                  |
| **Chol**       | Colesterol sérico en mg/dl.                                                                  |
| **FBS**        | Nivel de azúcar en sangre en ayunas. 1 = verdadero, 0 = falso.                               |
| **Restecg**    | Resultados electrocardiográficos en reposo.                                                  |
|            | - Valor 0: Normal.                                                                           |
|            | - Valor 1: Con anormalidad en la onda ST-T.                                                 |
|            | - Valor 2: Probable o definitiva hipertrofia ventricular izquierda.                          |
| **Thalach**    | Frecuencia cardíaca máxima alcanzada.                                                        |
| **Exang**      | Angina inducida por ejercicio. 1 = sí, 0 = no.                                               |
| **Oldpeak**    | Depresión del segmento ST inducida por el ejercicio relativa al reposo.                       |
| **Slope**      | La pendiente del segmento ST durante el ejercicio.                                            |
|            | - Valor 1: Ascendente.                                                                       |
|            | - Valor 2: Plano.                                                                            |
|            | - Valor 3: Descendente.                                                                     |
| **CA**         | Número de vasos principales (0-3) coloreados por fluoroscopía.                                |
| **Thal**       | Resultados de la prueba de thalium.                                                          |
|            | - Valor 3: Normal.                                                                           |
|            | - Valor 6: Defecto fijo.                                                                     |
|            | - Valor 7: Defecto reversible.                                                               |
| **Num**        | Diagnóstico de enfermedad cardíaca (estado angiográfico).                                     |
|            | - Valor 0: Menos del 50% de estrechamiento del diámetro en cualquier vaso importante.         |
|            | - Valor 1: Más del 50% de estrechamiento del diámetro en cualquier vaso importante.           |



###**Librerias utilizadas**

```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.io import arff
import matplotlib.pyplot as plt
import warnings
import os
warnings.filterwarnings('ignore')
!git clone https://github.com/javel3232/cholesterol.git
os.chdir('/content/cholesterol/Data')
```

###**Carga de los Datos**

```python
data, meta = arff.loadarff('dataset_2190_cholesterol.arff')
df = pd.DataFrame(data)
df.head(10)
```

###**Información de los datos**

```python
df.info()
display(df.describe())
print(f"Cantidad de filas en el archivo de entrenamiento:\t{df.shape[0]}")
print(f"Cantidad de columnas en el archivo de entrenamiento:\t{df.shape[1]}\n")
```

###**Valores Faltantes**

```python
display(pd.DataFrame({'Variables': df.isna().sum().index, 'Valores faltantes': df.isna().sum().values}))
```

###**Imputación de valores faltantes en la columna "ca" con la media**

```python
mean_ca = df['ca'].mean()
df['ca'].fillna(mean_ca, inplace=True)
display(pd.DataFrame({'Variables': df.isna().sum().index, 'Valores faltantes': df.isna().sum().values}))
```

###**Matriz de grafica de correlaciones**

```python
sns.set(style='whitegrid', context='notebook')
cols=['age','trestbps','thalach','oldpeak','chol']
sns.pairplot(df[cols] , height=2)
plt.show
```

###**Mapa de calor de correlación**

```python
# Mapa de Calor de Correlación
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor de Correlación")
plt.show()
display(pd.DataFrame(correlation_matrix))
```

**Edad (Age):**
Tiene una correlación positiva moderada con la presión arterial en reposo (trestbps) y el colesterol sérico (chol), lo que indica que a medida que la edad aumenta, estos valores tienden a aumentar también, aunque la relación no es muy fuerte.

Tiene una correlación negativa moderada con la frecuencia cardíaca máxima alcanzada (thalach), lo que sugiere que a medida que la edad aumenta, la frecuencia cardíaca máxima alcanzada tiende a disminuir.

**Presión arterial en reposo (Trestbps):**
Tiene una correlación positiva moderada con la edad y el colesterol sérico, lo que sugiere que a medida que la presión arterial en reposo aumenta, la edad y el colesterol sérico tienden a aumentar también, aunque la relación no es muy fuerte.

**Frecuencia cardíaca máxima alcanzada (Thalach):**
Tiene una correlación negativa moderada con la edad y la depresión del segmento ST inducida por el ejercicio relativa al reposo (oldpeak), lo que indica que a medida que la frecuencia cardíaca máxima alcanzada aumenta, la edad y la depresión del segmento ST inducida por el ejercicio tienden a disminuir.

**Depresión del segmento ST inducida por el ejercicio relativa al reposo (Oldpeak):**
Tiene una correlación positiva moderada con el número de vasos principales coloreados por fluoroscopía (CA), lo que sugiere que a medida que la depresión del segmento ST inducida por el ejercicio aumenta, el número de vasos principales coloreados por fluoroscopía también tiende a aumentar.

**Colesterol sérico (Chol):**
Tiene una correlación positiva débil con la edad y la presión arterial en reposo, lo que sugiere una relación ligeramente positiva entre estas variables.

```python
plt.figure(figsize=(8,6))
plt.title("Distribución de las variables en el conjunto de datos ")
for i, col in enumerate(cols, 1):
    plt.subplot(3, 2, i)
    plt.title(f"Distribución de los datos de {col}")
    sns.histplot(df[col], kde=True)
    plt.tight_layout()
    plt.plot()
plt.show()
```

###**Box plots**

```python
plt.figure(figsize=[12,10])

sns.set(style="whitegrid")

plt.subplot(231)
plt.boxplot(x=df['age'], showmeans=True, meanline=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Age')
plt.ylabel('Frequency')

plt.subplot(232)
plt.boxplot(x=df['trestbps'], showmeans=True, meanline=True, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Resting Blood Pressure (trestbps)')
plt.ylabel('Frequency')

plt.subplot(233)
plt.boxplot(x=df['thalach'], showmeans=True, meanline=True, patch_artist=True, boxprops=dict(facecolor='lightcoral'))
plt.title('Maximum Heart Rate Achieved (thalach)')
plt.ylabel('Frequency')

plt.subplot(234)
plt.boxplot(x=df['oldpeak'], showmeans=True, meanline=True, patch_artist=True, boxprops=dict(facecolor='lightgrey'))
plt.title('ST Depression Induced by Exercise (oldpeak)')
plt.ylabel('Frequency')

plt.subplot(235)
plt.boxplot(x=df['chol'], showmeans=True, meanline=True, patch_artist=True, boxprops=dict(facecolor='lightyellow'))
plt.title('Serum Cholesterol (chol)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
